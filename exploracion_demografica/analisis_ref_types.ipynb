```python
import pandas as pd
import numpy as np
import seaborn as sns

types_auctions = {'auction_type_id':'category','country':'int64',
         'device_id':'int64','platform':'category',
         'ref_type':'category','source_id':'category'}
auctions = pd.read_csv('/home/fedoremelianenko/FIUBA/Organizacion de datos/TP1/data/auctions.csv',
                       dtype=types_auctions,
                       usecols=['date','device_id','platform','ref_type_id','source_id'])
auctions.head()
```

```python
auctions_source_id_vc = auctions['source_id'].value_counts()
auctions_source_id_vc
```

```python
ax = auctions_source_id_vc.plot.pie(label='Fuente',autopct='%1.1f%%', figsize=(12,12) )
ax.set_title('Fuentes de procedencia de subastas', fontsize=20)
```

```python
colors = ['blue','orange','green','red','purple']
ax = auctions_source_id_vc.plot.bar(x='Fuente',y='Cantidad de auctions', figsize = (12,12),
                               color=colors)
ax.set_title('Auctions procedentes de cada fuente', fontsize=20)
ax.set_xlabel('Fuente',fontsize=18)
ax.set_ylabel('Numero de auctions',fontsize=18)
```

```python
colors_pie = ['lightgreen','lightblue']
auctions_ref_id_vc = auctions['platform'].value_counts()
ax = auctions_ref_id_vc.plot.pie(colors=colors_pie, explode=[0,0.2], figsize=(12,12),
                                 shadow=True, label = 'Plataforma', autopct='%1.1f%%')
ax.set_title('Fuentes de procedencia de subastas', fontsize=20)
```

```python
colors = ['lightgreen','lightblue']
ax = auctions_ref_id_vc.plot.bar(x='Fuente',y='Cantidad de auctions', color=colors, figsize =(12,12))
ax.set_xlabel('Plataforma',fontsize=16)
ax.set_ylabel('Cantidad de auctions',fontsize=16)
ax.set_title('Auctions procedentes de cada fuente',fontsize=18)
```
